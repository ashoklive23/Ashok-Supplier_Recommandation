```python
# Define criteria, subcriteria, and alternatives
criteria = ['Sayur', 'Packaging', 'Biji Kopi']
subcriteria_waktu = ['Cepat', 'Sedang', 'Lambat']
subcriteria_harga = ['Murah', 'Sedang', 'Mahal']
subcriteria_kualitas = ['Rendah', 'Sedang', 'Tinggi']
alternatives = ['Supplier A', 'Supplier B', 'Supplier C']

# Define pairwise comparison matrices for each criterion
pairwise_comparison_criteria = [
    np.array([[1, 3, 2], [1/3, 1, 1/2], [1/2, 2, 1]]),
    np.array([[1, 3, 1/2], [1/3, 1, 1/3], [2, 3, 1]]),
    np.array([[1, 1/5, 1/4], [5, 1, 2], [4, 1/2, 1]])
]

# Define pairwise comparison matrices for each sub-criterion
pairwise_comparison_subcriteria_waktu = [
    np.array([[1, 1/3, 1/5], [3, 1, 1/3], [5, 3, 1]]),
    np.array([[1, 1/3, 1/5], [3, 1, 1/3], [5, 3, 1]]),
    np.array([[1, 1/2, 1/3], [2, 1, 1/2], [3, 2, 1]])
]
pairwise_comparison_subcriteria_harga = [
    np.array([[1, 1/3, 1/5], [3, 1, 1/3], [5, 3, 1]]),
    np.array([[1, 1/3, 1/5], [3, 1, 1/3], [5, 3, 1]]),
    np.array([[1, 1/2, 1/3], [2, 1, 1/2], [3, 2, 1]])
]
pairwise_comparison_subcriteria_kualitas = [
    np.array([[1, 1/3, 1/5], [3, 1, 1/3], [5, 3, 1]]),
    np.array([[1, 1/3, 1/5], [3, 1, 1/3], [5, 3, 1]]),
    np.array([[1, 1/2, 1/3], [2, 1, 1/2], [3, 2, 1]])
]

# Define weights for each criterion
weights_criteria = np.array([0.4, 0.3, 0.3])

# Define weights for each sub-criterion
weights_subcriteria_waktu = np.array([0.6, 0.3, 0.1])
weights_subcriteria_harga = np.array([0.5, 0.3, 0.2])
weights_subcriteria_kualitas = np.array([0.4, 0.4, 0.2])

# Calculate the overall weights for sub-criteria under each criterion
overall_weights_subcriteria_waktu = weights_criteria[0] * weights_subcriteria_waktu
overall_weights_subcriteria_harga = weights_criteria[1] * weights_subcriteria_harga
overall_weights_subcriteria_kualitas = weights_criteria[2] * weights_subcriteria_kualitas

#Normalize the overall weights for sub-criteria under each criterion
overall_weights_subcriteria_waktu /= np.sum(overall_weights_subcriteria_waktu)
overall_weights_subcriteria_harga /= np.sum(overall_weights_subcriteria_harga)
overall_weights_subcriteria_kualitas /= np.sum(overall_weights_subcriteria_kualitas)

#Calculate the weighted sum of the sub-criteria for each alternative
weighted_sum_subcriteria_supplier_a = np.array([
np.sum(pairwise_comparison_subcriteria_waktu[0] * overall_weights_subcriteria_waktu),
np.sum(pairwise_comparison_subcriteria_harga[0] * overall_weights_subcriteria_harga),
np.sum(pairwise_comparison_subcriteria_kualitas[0] * overall_weights_subcriteria_kualitas)
])
weighted_sum_subcriteria_supplier_b = np.array([
np.sum(pairwise_comparison_subcriteria_waktu[1] * overall_weights_subcriteria_waktu),
np.sum(pairwise_comparison_subcriteria_harga[1] * overall_weights_subcriteria_harga),
np.sum(pairwise_comparison_subcriteria_kualitas[1] * overall_weights_subcriteria_kualitas)
])
weighted_sum_subcriteria_supplier_c = np.array([
np.sum(pairwise_comparison_subcriteria_waktu[2] * overall_weights_subcriteria_waktu),
np.sum(pairwise_comparison_subcriteria_harga[2] * overall_weights_subcriteria_harga),
np.sum(pairwise_comparison_subcriteria_kualitas[2] * overall_weights_subcriteria_kualitas)
])

#Calculate the weighted sum of the criteria for each alternative
weighted_sum_criteria_supplier_a = np.sum(weighted_sum_subcriteria_supplier_a * weights_criteria)
weighted_sum_criteria_supplier_b = np.sum(weighted_sum_subcriteria_supplier_b * weights_criteria)
weighted_sum_criteria_supplier_c = np.sum(weighted_sum_subcriteria_supplier_c * weights_criteria)

#Print the results
print(f"Weighted sum of criteria for Supplier A: {weighted_sum_criteria_supplier_a:.4f}")
print(f"Weighted sum of criteria for Supplier B: {weighted_sum_criteria_supplier_b:.4f}")
print(f"Weighted sum of criteria for Supplier C: {weighted_sum_criteria_supplier_c:.4f}")
```
